Converting datatypes

````python
apps['Reviews']
````

````python
# Check current datatype of Reviews column
apps['Reviews'].dtype
````

````python
# Converting 'Reviews' to integer datatype
apps['Reviews'] = apps['Reviews'].astype(int)
````

````python
# Verifying the Conversion
apps['Reviews'].dtype
````

````python
# Installs column has extra symbols like '+' and ','
apps['Installs']
````

````python
# Checking 'Installs' datatype
apps['Installs'].dtype
````

````python
# Removing '+' and ',' symbols
apps['Installs'] = apps['Installs'].str.replace('+', '', regex=False)
apps['Installs'] = apps['Installs'].str.replace(',', '', regex=False)
````

````python
# Convert to numeric
apps['Installs'] = apps['Installs'].astype(int)
````

````python
# Verify the conversion
apps['Installs'].dtype
````

````python
# Price column has '$' symbol
apps['Price'].unique()
````

````python
# "$" symbol needs to be removed
apps['Price'] = apps['Price'].str.replace('$', '', regex=False)
````

````python
# Convert to float datatype
apps['Price'] = apps['Price'].astype(float)
````

````python
# Verify the conversion
apps['Price'].dtype
````

````python
# Size column
apps['Size']
````

````python
# Convert sizes 'k' to 'MB'
# 1 MB = 1024 kb
apps['Size'] = apps['Size'].apply(
    lambda x: float(x.replace('k', '')) / 1024 if isinstance(x, str) and 'k' in x else x
)
````

````python
# Convert 'varies with devices' to NaN as this value cannot be converted to a number, so we convert it to NaN
apps['Size'] = apps['Size'].replace('Varies with device', np.nan)
````

````python
# Remove 'M' and convert to float
apps['Size'] = apps['Size'].apply(
    lambda x: float(x.replace('M', '')) if isinstance(x, str) and 'M' in x else x
)
````

````python
# Check datatype
apps['Size'].dtype
````

````python
# Handling NaN values in size column by filling with median
apps['Size'].fillna(apps['Size'].median(), inplace=True)
````

````python
# Last Updated Column
apps['Last Updated']
````

````python
# Check datatype of Last Updated
apps['Last Updated'].dtype
````

````python
# Convert to datetime
apps['Last Updated'] = pd.to_datetime(apps['Last Updated'], errors='coerce')
````

````python
# Verify conversion
apps['Last Updated'].dtype
````

````python
# We saw a rating value around 19, which is impossible because rating value ranges between 1 to 5.
````

````python
# Confirm the outlier
apps[apps['Rating'] > 5]
````

````python
# Fix the outlier by removing the row
apps = apps[apps['Rating'] <= 5]
````

````python
# Verify the Fix
apps['Rating'].max()
````
